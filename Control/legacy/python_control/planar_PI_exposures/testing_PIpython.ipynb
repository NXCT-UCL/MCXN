```python
import pipython

PI_COM = 3
PI_BR = 9600

# Create a connection to the PI controller
pidevice = pipython.GCSDevice() 

# Open the connection
pidevice.ConnectRS232(PI_COM, PI_BR)

pidevice.MOV(3, 0)
print(pidevice.qPOS())

pidevice.CloseConnection()
```

```python
import pipython

# Create a connection to the PI controller
controller = pipython.GCSDevice('COM3')  # Replace 'COM3' with the actual serial port

# Open the connection
controller.ConnectRS232(3, 9600)

# Define the command to move the X-axis (replace with the actual command)
x_axis_move_command = "MOV X 10"  # Move the X-axis by 10 units (replace with your desired value)

# Send the command to the controller
controller.send(x_axis_move_command)

# Wait for a response (if needed)
response = controller.qIDN()  # Replace with an appropriate query command
print(f"Controller response: {response}")

# Close the connection
controller.CloseConnection()
```

```python
from pipython.pidevice.gcscommands import GCSCommands
from pipython.pidevice.gcsmessages import GCSMessages
from pipython.pidevice.interfaces.piserial import PISerial
pidevice = GCSCommands(messages = GCSMessages(PISerial(port='COM3', baudrate=9600)))
```

```python
from pipython.pidevice.gcscommands import GCSCommands
from pipython.pidevice.gcsmessages import GCSMessages
from pipython.pidevice.interfaces.piserial import PISerial
with PISerial(port='COM3', baudrate=9600) as gateway:
    messages = GCSMessages(gateway)
    pidevice = GCSCommands(messages)
    print(pidevice.qIDN())
```
